## 1. Initial Setup: Importing Libraries and Loading Data

We also import `imblearn` to handle the class imbalance in our dataset.

## 3. Feature Engineering and Preprocessing

### 3.1 Data Cleaning and Feature Creation
Here, we'll create new features from the existing data to provide more predictive signals to our model. We will also drop columns that are not useful.

```python
def feature_engineer(df):
    """Cleans the dataframe and creates new features."""
    
    # Drop identifier columns
    df = df.drop(columns=['Unnamed: 0', 'cc_num', 'first', 'last', 'street', 'city', 'state', 'zip', 'job', 'trans_num', 'unix_time'])
    
    # Convert date columns to datetime
    df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
    df['dob'] = pd.to_datetime(df['dob'])
    
    df['hour'] = df['trans_date_trans_time'].dt.hour
    df['day_of_week'] = df['trans_date_trans_time'].dt.dayofweek # Monday=0, Sunday=6
    df['month'] = df['trans_date_trans_time'].dt.month
    df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)
    
    # Age
    df['age'] = (df['trans_date_trans_time'].dt.year - df['dob'].dt.year)

    # Earth radius in kilometers
    R = 6371
    
    # Convert degrees to radians
    lat1_rad = np.radians(df['lat'])
    lon1_rad = np.radians(df['long'])
    lat2_rad = np.radians(df['merch_lat'])
    lon2_rad = np.radians(df['merch_long'])
    
    # Differences in coordinates
    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad
    
    # Haversine formula
    a = np.sin(dlat / 2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance = R * c
    
    df['distance'] = distance
    
    # Drop original columns that are now redundant
    df = df.drop(columns=['trans_date_trans_time', 'dob', 'lat', 'long', 'merch_lat', 'merch_long'])
    
    return df

print("Applying feature engineering...")
df_engineered = feature_engineer(df_full.copy())
print("Feature engineering complete!")
display(df_engineered.head())
```

### 3.2 Splitting Data and Setting Up the Preprocessing Pipeline
Now we split the data back into training and testing sets and define our preprocessing steps. We will:
1.  **Scale Numerical Features:** Use `StandardScaler` to normalize features like `amt`, `age`, and `distance`. This is crucial for models like Logistic Regression and helps gradient-based models converge faster.
2.  **Encode Categorical Features:** Use `OneHotEncoder` to convert categorical columns like `category` and `gender` into a numerical format that the model can understand.

```python
# Separate features (X) and target (y)
X = df_engineered.drop('is_fraud', axis=1)
y = df_engineered['is_fraud']

# Split back into train and test sets using the stored index
X_train = X.iloc[:train_set_size]
X_test = X.iloc[train_set_size:]
y_train = y.iloc[:train_set_size]
y_test = y.iloc[train_set_size:]

print(f"Training set dimensions: {X_train.shape}")
print(f"Testing set dimensions: {X_test.shape}")
```

```python
# Identify column types for the pipeline
numerical_features = X_train.select_dtypes(include=np.number).columns.tolist()
categorical_features = X_train.select_dtypes(exclude=np.number).columns.tolist()

print(f"Numerical features to scale: {numerical_features}")
print(f"Categorical features to encode: {categorical_features}")

# Create the preprocessing pipeline using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough' # In case we have columns we didn't specify
)
```

--- 
## 4. Model Training and Evaluation

### 4.1 Building the Model Pipelines with SMOTE
To handle the class imbalance, we will integrate `SMOTE` directly into our model pipelines. This is crucial because **SMOTE should only be applied to the training data**. An `ImbPipeline` ensures that during cross-validation, SMOTE is applied correctly to each fold without data leakage.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, fbeta_score

f2_scorer = make_scorer(fbeta_score, beta=2)

param_grids = {
    'Logistic Regression': {
        'classifier__C': [0.1, 1.0, 10],
        'classifier__penalty': ['l1', 'l2']
    },
    'Random Forest': {
        'classifier__n_estimators': [100, 150],
        'classifier__max_depth': [10, 15],
    },
    'XGBoost': {
        'classifier__n_estimators': [100, 150],
        'classifier__learning_rate': [0.05, 0.1],
        'classifier__max_depth': [5, 7],
        'classifier__eval_metric': ['aucpr'] # Focused on the better metric
    },
    'LightGBM': {
        'classifier__n_estimators': [100, 150],
        'classifier__learning_rate': [0.05, 0.1],
        'classifier__num_leaves': [20, 31]
    }
}

models = {
    'Logistic Regression': LogisticRegression(random_state=42, solver='liblinear'),
    'Random Forest': RandomForestClassifier(random_state=42), # n_jobs removed
    'XGBoost': XGBClassifier(random_state=42, use_label_encoder=False),
    'LightGBM': LGBMClassifier(random_state=42) # n_jobs removed
}

best_models = {}
for name, model in models.items():
    print(f"--- Tuning {name} ---")
    
    pipeline = ImbPipeline([
        ('preprocessor', preprocessor),
        ('smote', SMOTE(random_state=42)),
        ('classifier', model)
    ])
    grid_search = GridSearchCV(
        estimator=pipeline,
        param_grid=param_grids[name],
        cv=3,
        scoring=f2_scorer,
        n_jobs=2, # Reduced from -1 to 2 to save memory
        verbose=1
    )
    
    grid_search.fit(X_train, y_train)
    
    best_models[name] = grid_search.best_estimator_
    print(f"Best parameters for {name}: {grid_search.best_params_}")
    print(f"Best F2 score for {name}: {grid_search.best_score_:.4f}\n")

print("Hyperparameter tuning complete.")
```

### 4.2 Evaluating Model Performance
Now we evaluate our trained models on the **unseen test data**. We'll create a helper function to display the key metrics and the confusion matrix for each model.

```python
def evaluate_model(model, X_test, y_test, model_name):
    
    # Make predictions
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    
    # Metrics
    print(f"\n--- Evaluation for {model_name} ---")
    print(classification_report(y_test, y_pred, target_names=['Not Fraud', 'Fraud']))
    pr_auc = average_precision_score(y_test, y_proba)
    print(f"Precision-Recall (PR) AUC Score: {pr_auc:.4f}")

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Predicted Not Fraud', 'Predicted Fraud'], 
                yticklabels=['Actual Not Fraud', 'Actual Fraud'])
    plt.title(f'{model_name} Confusion Matrix')
    plt.show()

for name, model in best_models.items():
    evaluate_model(model, X_test, y_test, name)
```

### 4.3 Interpretation of Results

* **Logistic Regression:** Serves as a good baseline. Its recall for the 'Fraud' class is decent, but it likely misclassifies many non-fraudulent transactions as fraud (lower precision).

* **XGBoost & LightGBM:** These models significantly outperform the baseline. 
    * **Recall (Sensitivity):** This is often the most important metric for fraud detection. It answers: *"Of all the actual fraudulent transactions, what percentage did we correctly identify?"* A high recall is critical to minimize financial losses. Both XGBoost and LightGBM show very high recall.
    * **Precision:** This answers: *"Of all the transactions we flagged as fraud, what percentage were actually fraudulent?"* High precision is important to avoid inconveniencing legitimate customers with false alarms.
    * **F1-Score:** The harmonic mean of precision and recall, providing a single score that balances both concerns.
    * **PR AUC:** This is a very reliable metric for imbalanced datasets, summarizing the trade-off between precision and recall across different probability thresholds.

**Conclusion:** Both **XGBoost** and **LightGBM** are excellent candidates. LightGBM is often faster, while XGBoost can sometimes be slightly more accurate (not here). Based on the metrics, they both demonstrate a strong ability to identify fraud while maintaining a reasonable level of precision.
