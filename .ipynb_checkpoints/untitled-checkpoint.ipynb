# Advanced Credit Card Fraud Detection

* **Exploratory Data Analysis (EDA):** Deep dives into transaction patterns.
* **Advanced Models:** Implementation of XGBoost and LightGBM, known for high performance on tabular data.
* **Hyperparameter Tuning:** Using `RandomizedSearchCV` to find optimal model parameters.
* **Feature Importance:** Visualizing which features contribute most to the model's predictions.

## 1. Setup and Data Loading

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Scikit-learn imports
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_recall_curve

# Imbalanced-learn import
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

# Advanced model imports
import xgboost as xgb
import lightgbm as lgb

# Settings
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
```

```python
# Load the datasets
try:
    df_train = pd.read_csv('fraudTrain.csv')
    df_test = pd.read_csv('fraudTest.csv')
    print("Data loaded successfully.")
except FileNotFoundError:
    print("ERROR: Make sure 'fraudTrain.csv' and 'fraudTest.csv' are in the same directory.")
    # Create empty dataframes if files are not found to avoid further errors
    df_train = pd.DataFrame()
    df_test = pd.DataFrame()

# Combine for preprocessing
if not df_train.empty and not df_test.empty:
    df = pd.concat([df_train, df_test], ignore_index=True)
    print(f"Total transactions: {len(df)}")
    print(f"Data shape: {df.shape}")
```

## 2. Preprocessing and Feature Engineering

```python
def preprocess_data(df):
    """Cleans and engineers features for the fraud detection dataset."""
    # Drop unnecessary columns
    df = df.drop(['Unnamed: 0', 'trans_num', 'cc_num', 'first', 'last', 'street', 'city', 'zip', 'unix_time'], axis=1)
    
    # Convert date columns to datetime objects
    df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
    df['dob'] = pd.to_datetime(df['dob'])
    
    # --- Feature Engineering ---
    # Time-based features
    df['hour'] = df['trans_date_trans_time'].dt.hour
    df['day_of_week'] = df['trans_date_trans_time'].dt.dayofweek
    df['month'] = df['trans_date_trans_time'].dt.month
    
    # Age of cardholder at time of transaction
    df['age'] = (df['trans_date_trans_time'].dt.year - df['dob'].dt.year)
    
    # Interaction features (example)
    df['amt_per_age'] = df['amt'] / (df['age'] + 1) # Add 1 to avoid division by zero
    
    # Drop original date columns
    df = df.drop(['trans_date_trans_time', 'dob'], axis=1)
    
    return df

print("Applying preprocessing and feature engineering...")
df_processed = preprocess_data(df.copy())
print("Preprocessing complete.")
df_processed.head()
```

## 3. Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='is_fraud', data=df_processed)
plt.title('Distribution of Fraud vs. Non-Fraud Transactions')
plt.show()

print(df_processed['is_fraud'].value_counts(normalize=True))
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df_processed, x='hour', hue='is_fraud', common_norm=False, kde=True, palette='viridis')
plt.title('Transaction Distribution by Hour')
plt.show()
```

## 4. Data Splitting and Pipeline Setup

```python
# Define features (X) and target (y)
X = df_processed.drop('is_fraud', axis=1)
y = df_processed['is_fraud']

# Split back into original train and test sets based on original file lengths
train_size = len(df_train)
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]

print(f"Training set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")
```

```python
# Identify numerical and categorical columns
numerical_cols = X_train.select_dtypes(include=np.number).columns.tolist()
categorical_cols = X_train.select_dtypes(exclude=np.number).columns.tolist()

# Create a preprocessor object
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough'
)
```

## 5. Model Training: XGBoost and LightGBM

```python
# --- XGBoost Pipeline ---
xgb_pipeline = ImbPipeline([
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('classifier', xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'))
])

print("Training XGBoost model...")
xgb_pipeline.fit(X_train, y_train)
print("XGBoost training complete.")
```

```python
# --- LightGBM Pipeline ---
lgb_pipeline = ImbPipeline([
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('classifier', lgb.LGBMClassifier(random_state=42))
])

print("\nTraining LightGBM model...")
lgb_pipeline.fit(X_train, y_train)
print("LightGBM training complete.")
```

## 6. Model Evaluation

```python
def evaluate_model(pipeline, X_test, y_test, model_name):
    """Generates and prints evaluation metrics for a given model pipeline."""
    print(f"--- {model_name} Evaluation ---")
    y_pred = pipeline.predict(X_test)
    y_proba = pipeline.predict_proba(X_test)[:, 1]
    
    # Classification Report
    print("Classification Report:")
    print(classification_report(y_test, y_pred, target_names=['Not Fraud', 'Fraud']))
    
    # ROC AUC Score
    print(f"ROC AUC Score: {roc_auc_score(y_test, y_proba):.4f}\n")
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
    plt.title(f'{model_name} Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

evaluate_model(xgb_pipeline, X_test, y_test, "XGBoost")
evaluate_model(lgb_pipeline, X_test, y_test, "LightGBM")
```

## 7. Feature Importance (XGBoost)

Let's see which features our best model (XGBoost) found most important.

```python
# Get feature names after one-hot encoding
try:
    ohe_feature_names = xgb_pipeline.named_steps['preprocessor'].named_transformers_['cat'].get_feature_names_out(categorical_cols).tolist()
    final_feature_names = numerical_cols + ohe_feature_names
    
    # Get feature importances from the trained XGBoost model
    importances = xgb_pipeline.named_steps['classifier'].feature_importances_
    
    # Create a DataFrame for visualization
    importance_df = pd.DataFrame({'feature': final_feature_names, 'importance': importances})
    importance_df = importance_df.sort_values('importance', ascending=False).head(20) # Top 20 features
    
    # Plot
    plt.figure(figsize=(12, 8))
    sns.barplot(x='importance', y='feature', data=importance_df)
    plt.title('Top 20 Feature Importances (XGBoost)')
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Could not plot feature importance: {e}")
```
